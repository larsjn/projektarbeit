```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
from ipywidgets import Layout
from ipywidgets import Box
from ipywidgets import Button
from ipywidgets import Text
from ipywidgets import IntSlider
from ipywidgets import Dropdown
from ipywidgets import IntText
from ipywidgets import FloatText


import sys
```

```python
class Create_New_Function(object) :
    
    def __init__(self):     
        # ====================================== Definition der Angezeigten Texte ====================================== 
        # Acc Titel
        self.RS_Acc_Titel_Setup = 'Grundeinstellungen der Funktion'
        self.RS_Acc_Titel_Mask = 'Eingabe der Funktion'
        
        # Texte 
        self.RS_Funktion='Funktion:'
        self.RS_Anzahl='Anzahl:'
        self.RS_Nichts = ''
        self.RS_Titel_Setup_Funktion = 'Anzahl der Teil Funktionen und Operatoren'
        self.RS_Titel_Typ_Funktion = 'Typ der Funktion'
        self.RS_Titel_Mask_Funktion = 'Eingabe der Funktion: Hinweis Variable mit "x" eintragen'
        
        self.RS_Zahl = 'Zahl' 

        # Help Texte
        self.RS_Help_Class = 'Gibt Erzeugt mittels "Create_fkt_layout" die Eingabemaske zum erzeugen einer neuen Funktion'
        self.RS_Help_Setup_Funktion = '1: Widgets der Eingabemaske der ANZ, 2: Eingabemaske ANZ, 3: Widgets Eingabemaste Setup, 4: Eingabemaste Setup'

        # Button Texte
        self.RS_Btn_Eingabemaske_erzeugen = 'Eingabemaske erzeugen'
        self.RS_Btn_loeschen = 'Fenster löschen'
        self.RS_Btn_save = 'Speichern'
        # Dropdown Menu Einträge
        self.RS_Dpd_Reelle_Funktion = 'Reelle Funktion'

        self.RS_Dpd_Plus = '+'
        self.RS_Dpd_Minus = '-'
        self.RS_Dpd_Mul = '*'
        self.RS_Dpd_Div = '/'
        self.RS_Dpd_Sin = 'Sin('
        self.RS_Dpd_Cos = 'Cos('
        self.RS_Dpd_Klammer_auf = '('
        self.RS_Dpd_Klammer_zu =')'       
        # ====================================== Definition "def" übergreifende Variablen ====================================== 
        self.FMaskResultIteams = None
        self.FMaskResultForm = None
        # ====================================== Definition "def" übergreifende Widgets ====================================== 
        self.F_wdg_Layout = Layout(display='flex',flex_flow='row',justify_content='center')
        self.F_wdg_Box_Layout = Layout(display='flex',flex_flow='column',border='solid 2px',align_items='stretch',width='50%')
        
        self.Fsld_Number_Sub_Fkt = IntSlider(layout=self.F_wdg_Layout,value=10,min=1,max=20,step=1,description='Anzahl:',
                                   disabled=False,continuous_update=False,slider_color='white')
        self.FAccItems = []
        self.FAcc = None
     
        # ====================================== Widget Modul Namen ======================================
        
        self.Modul_Name_FloatTextModel = 'FloatTextModel'
        self.Modul_Name_DropdownModel ='DropdownModel'
        
        
        # ====================================== Definition erstellen ====================================== 

    def help(self):       
        print(self.RS_Help_Class)
    
    def case_Plus(self):
         print(self.RS_Dpd_Plus) 
    def case_Minus(self):
        print(self.RS_Dpd_Minus) 
    def case_Mul(self):
        print(self.RS_Dpd_Mul) 
    def case_Div(self):
        print(self.RS_Dpd_Div) 
    def case_Sin(self):
        print(self.RS_Dpd_Sin) 
    def case_Cos(self):
        print(self.RS_Dpd_Cos) 
    def case_Klammer_auf(self):
        print(self.RS_Dpd_Klammer_auf) 
    def case_Klammer_zu(self):
        print(self.RS_Dpd_Klammer_zu) 
    
        
        
# ---------------------------------------------------- End def -----------------------------------------------------  
    def Btn_Event_Create_fkt_mask(self,ADummy): # Erstellt die Eingabemaske der Funktion
#         print("Debug Ausgabe")
        
        tmp1 = [Text(layout=self.F_wdg_Layout,value=self.RS_Titel_Mask_Funktion, disabled=True)]

        tmp2 = [FloatText(Layout=self.F_wdg_Layout,
        value=1,
        description=self.RS_Zahl,
        disabled=False
        )for i in range(self.Fsld_Number_Sub_Fkt.value)]
        
        tmp3= [Dropdown(Layout=self.F_wdg_Layout,
                       options = [self.RS_Dpd_Plus ,
                                  self.RS_Dpd_Minus,
                                  self.RS_Dpd_Mul,
                                  self.RS_Dpd_Div,
                                  self.RS_Dpd_Sin,
                                  self.RS_Dpd_Cos,
                                  self.RS_Dpd_Klammer_auf,
                                  self.RS_Dpd_Klammer_zu
                                  ], value=self.RS_Dpd_Plus ,description='Funktion:',
                                    disabled=False, button_style='' # 'success', 'info', 'warning', 'danger' or ''
                                    ) for i in range(self.Fsld_Number_Sub_Fkt.value)]

        for i in range(self.Fsld_Number_Sub_Fkt.value):    
            tmp1 =  tmp1+[tmp2[i]]+[tmp3[i]]

        btn_save=  Button(layout=self.F_wdg_Layout,button_style='warning',description=self.RS_Btn_save)
        btn_save.on_click(self.Btn_Event_Convert_to_Fkt) 
            
        form_fkt_mask_items = tmp1 + [btn_save]     

        form_fkt_mask = Box(form_fkt_mask_items , layout=self.F_wdg_Box_Layout)            
        self.FMaskResultIteams = form_fkt_mask_items
        self.FMaskResultForm = form_fkt_mask
        if len(self.FAccItems) > 1:
            self.FAccItems[1].close           
        self.FAccItems = [self.FAccItems[0]] + [ form_fkt_mask ]
        self.FAcc.children = self.FAccItems    
        
# ---------------------------------------------------- End def -----------------------------------------------------    
    def Btn_Event_Convert_to_Fkt(self,ADummy):
        print("Debug Ausgabe")  
        switch = { self.RS_Dpd_Plus : self.case_Plus,
                   self.RS_Dpd_Minus : self.case_Minus,
                   self.RS_Dpd_Mul : self.case_Mul,
                   self.RS_Dpd_Div : self.case_Div,
                   self.RS_Dpd_Sin : self.case_Sin,
                   self.RS_Dpd_Cos : self.case_Cos,
                   self.RS_Dpd_Klammer_auf : self.case_Klammer_auf,
                   self.RS_Dpd_Klammer_zu : self.case_Klammer_zu,
                } 
        for i in self.FMaskResultIteams:
                if i._model_name == self.Modul_Name_FloatTextModel : 
                    print(i._model_name)
                    
                if i._model_name == self.Modul_Name_DropdownModel :
                    switch[i.value]()
                
                
            
#         
# ---------------------------------------------------- End def -----------------------------------------------------

        
    def Create_fkt_layout(self,ADummy): # Erstellt die Eingabemöglichkeit für die Anzahl der SubFunktionen
       
        b = Button(layout=self.F_wdg_Layout,button_style='warning',description=self.RS_Btn_Eingabemaske_erzeugen)
        b.on_click(self.Btn_Event_Create_fkt_mask)

        form_fkt_items = [
                Text(layout=self.F_wdg_Layout, value=self.RS_Titel_Setup_Funktion, disabled=True),

                self.Fsld_Number_Sub_Fkt,   


                Text(layout= self.F_wdg_Layout, value=self.RS_Titel_Typ_Funktion, disabled=True),

                Dropdown(options=[self.RS_Dpd_Reelle_Funktion, '', ''],value=self.RS_Dpd_Reelle_Funktion,description=self.RS_Funktion,disabled=False,
                         button_style='' # 'success', 'info', 'warning', 'danger' or ''
                        ),

                 b,
        ] 

        form_new_fkt = Box(form_fkt_items, layout=self.F_wdg_Box_Layout)
        if len(self.FAccItems) > 0:
            self.FAccItems[0].close
        self.FAccItems = [form_new_fkt]
        self.FAcc = widgets.Accordion(children=self.FAccItems)
        self.FAcc.set_title(0, self.RS_Acc_Titel_Setup)
        self.FAcc.set_title(1, self.RS_Acc_Titel_Mask)    
        display(self.FAcc)
        return [self.RS_Help_Setup_Funktion,form_fkt_items,form_new_fkt,self.FMaskResultIteams,self.FMaskResultForm]       
# ---------------------------------------------------- End def -----------------------------------------------------        
        
        
        
```
